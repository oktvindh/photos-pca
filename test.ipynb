```python
#ubah citra yang berbentuk matriks menjaid vector berdimensi 12000
import cv2
import numpy as np
import glob

foto = glob.glob('gabung/*')
foto_mtr = np.array([np.array(cv2.imread(i,0)) for i in foto]).astype(float)

print(foto_mtr)
```

```python
#Gabungkan data tersebut menjadi matriks data yang berukuran 50x12000
data = foto_mtr.reshape(50, 12000)
print(data.shape)
```

```python
#Hitung matriks koragamannya
data_cov = np.cov(data)
print(data_cov)
```

```python
#Hitung vector ciri dan akar ciri dari matriks koragam tersebut
akarCiri, vektorCiri = np.linalg.eig(data_cov)
print(akarCiri)
print(vektorCiri)
```

```python
#Tentukan banyaknya vector ciri yang akan diambil jika diinginkan proporsi informasi
# yang ditangkap oleh sumbu-sumbu baru adalah 90%
from sklearn.decomposition import PCA
data_pca = PCA(0.9)
data_pca.fit(data)
q_count = data_pca.n_components_
print(f"q yang diambil = {q_count}")
```

```python
#Lakukan transformasi data asli yang berukuran 50x12000 menjadi data hasil transformasi yang berukuran 50xg (q adalah banyaknya sumbu baru yang dihasilkan
# dari point 8)
data_trans = data_pca.transform(data)
print(data_trans)
```

```python
#Lakukan analisis klaster berdasar data hasil transformasi tersebut. Perhatikan
# anggota-anggota klaster jika diambil 10 klaster
from sklearn.cluster import KMeans
data_klaster = KMeans(n_clusters=10).fit(data_trans)
print(data_klaster.cluster_centers_)
```
